```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
iris= load_iris()
```

```python
X = iris.data
y = iris.target
```

```python
df = pd.DataFrame(np.c_[X, y], columns=iris.feature_names + ["Species"])
```

```python
df
```

```python
df.Species = df.Species.apply(lambda x: iris.target_names[int(x)])
```

```python
df.groupby('Species')[df.columns[:-1]].mean()
```

```python
sns.scatterplot(data=df, hue="Species", x="sepal width (cm)", y="sepal length (cm)")
plt.show()
sns.scatterplot(data=df, hue="Species", x="petal width (cm)", y="petal length (cm)", palette="icefire_r");
```
